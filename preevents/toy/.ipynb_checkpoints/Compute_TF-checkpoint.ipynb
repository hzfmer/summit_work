```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import scipy
from scipy.fftpack import fft, ifft
from scipy.signal import butter, sosfiltfilt, sosfilt, resample, detrend
from obspy.signal.konnoohmachismoothing import calculate_smoothing_matrix
from obspy.signal.konnoohmachismoothing import apply_smoothing_matrix
from obspy.signal.konnoohmachismoothing import konno_ohmachi_smoothing
from obspy.signal.detrend import polynomial
from matplotlib.ticker import LogFormatter, ScalarFormatter

%load_ext autoreload
%autoreload 2
plt.style.use(['science', 'ieee', 'high-vis', 'no-latex'])
```

```python
def filt_B(data_in, fs, lowcut=0, highcut=11.2, order=5, causal=False):
    """Butterworth filtering subroutine
    Parameters
    ----------
    data_in : array_like
        Input time series, can be multi-dimensional
    fs : float
        Sampling frequency
    lowcut : float
        Left end of band range. 0 means low-pass filtering
    highcut : float
        Right end of band range.
    order : int
        The order of the filter
    causal : bool
        Causal filter or acausal
        
    Outputs
    -------
    data_out : array_like
        Output with the same shape as data_in
    """
    if highcut >= fs / 2:
        return data_in
    sz = data_in.shape
    if len(sz) > 1 and sz[0] > sz[1]:
        data_in = data_in.T
    nyq = fs / 2
    low = lowcut / nyq
    high = highcut / nyq
    if low == 0:  # lowpass
        sos = butter(order, high, analog=False, btype="low", output="sos")
    else:  # bandpass
        sos = butter(order, [low, high], analog=False, btype="band", output="sos")
    if causal:
        data_out = sosfilt(sos, data_in)
        data_out = sosfilt(sos, data_out)
    else:
        data_out = sosfiltfilt(sos, data_in)
    if len(sz) > 1 and sz[0] > sz[1]:
        data_out = data_out.T
    return data_out
```

```python
def plot_set(ax, ymax, ymin=0.5, xmin=0.5, xmax=12):
    """Helper function for plotting
    """
    ax.set_xscale("log")
    ax.set_yscale("log")
    ax.set_xlim((xmin, xmax))
    ax.set_ylim((ymin, ymax))
    formatter = LogFormatter(labelOnlyBase=False, minor_thresholds=(10, 0.4))
    ax.get_xaxis().set_minor_formatter(formatter)
    ax.set_xticks([1])
    plt.minorticks_on()
    plt.grid(which="both", linestyle="--", alpha=0.5)
    ax.legend(loc=2)
    ax.set_ylabel("Amplification")
    ax.set_xlabel("Frequency (Hz)")
```

```python
def taper_vel(vel, dt, taper_seconds=0, pad_seconds=5):
    """
    This function tapers the last n seconds of vel using a hanning window. Pad pad_seconds zeros at the end
    """
    npts = int(2 * taper_seconds / dt +1)
    npad = int(pad_seconds / dt)
    window = np.hanning(npts)[-int(taper_seconds / dt + 1):]
    if len(vel.shape) > 1:
        for i in range(vel.shape[1]):      
            vel[-len(window):, i] *= window
    else:
        vel[-len(window):] *= window  # taper
    print('last 2 velocities: ', vel[-2:,:])
    vel = np.pad(vel, ((0, npad), (0, 0)))  # Pad zeros
    return vel
```

```python
def process_vel(vel, dt, high_cut, lowcut=0.5, filt=False,
                taper=0, detrend=False, baseline=False):
    """Processing tool
    
    The tool reads in velocities, and integrate to displacements, then output processed acceleration spectra.
    """
    if detrend:  # do detrend
        vel = scipy.signal.detrend(vel, axis=0)
        
    if taper > 0:  # tapering the last seconds of vel, appending zeros
        vel = taper_vel(vel, dt, taper_seconds=taper)
        
    if filt:  # Band-pass filtering vel
        vel = filt_B(vel, 1 / dt, lowcut=lowcut, highcut=high_cut)  # band-pass filter
    
    disp = np.cumsum(vel, axis=0) * dt
    if baseline:  # 2-nd order polynomial detrend on displacements      
        if len(vel.shape) > 1:  # do baseline correction on each component, if input has >1 components
            for i in range(vel.shape[-1]):
                disp[:, i] = polynomial(disp[:, i], order=2)
        else:  # baseline correction
            disp = polynomial(disp, order=2)
    # dis -> acc
    res = np.gradient(np.gradient(disp, dt, axis=0), dt, axis=0)    
    res = np.abs(fft(res, axis=0) * dt)  # fft  
    return res
```

```python
def comp_ttf(model, dt, f_lim, sm, filt=False, taper=0, detrend=False, baseline=False):
    """Computer the smoothed TTF"""
    ix, iy = 4, 2  # index of the site
    
    vx = np.fromfile(f"../data/vx_{model}.bin", "f").reshape((-1, 2, 4, 8))  # dt = 0.005s
    vy = np.fromfile(f"../data/vy_{model}.bin", "f").reshape((-1, 2, 4, 8))  # dt = 0.005s

    vx = vx[::2, :, iy, ix]  # every other step of t, so that dt increases from 0.005 to 0.01
    vy = vy[::2, :, iy, ix]  # shape(15000, 2)
    fx = process_vel(vx, dt, f_lim, filt=filt, taper=taper, detrend=detrend, baseline=baseline)
    fy = process_vel(vy, dt, f_lim, filt=filt, taper=taper, detrend=detrend, baseline=baseline)
    fa = np.sqrt(fx ** 2 + fy ** 2)  # spectra SRM
    res = fa[1: len(sm) + 1, 0] / fa[1 : len(sm) + 1, 1]  # spectrum ratio, take first half
    res[np.isnan(res)] = 1  # avoid possible divided-by-zero error
    res[np.isinf(res)] = 1
    res = apply_smoothing_matrix(res, sm, 1)  # smoothing
    
    return res
```

```python
def plot_ttf(event, dt, f_lim, filt=True, taper=False):
    etf_res = comp_ttf(f"{event}", dt, f_lim, filt=filt, taper=taper)
    fig, ax = plt.subplots(dpi=600)
    ax.fill_between(f_ref, etf_low, etf_up, color="gray", alpha=0.3, label="ETF 95% CI")
    ax.plot(f_ref, etf_res, label="TTF 3D")
    ax.plot(f_ref, ttf_1d_med, label="TTF 1D")
    plot_set(ax, 2 * etf_up.max())
```

Our procedure for processing the data is similar to the that documented in Tau and Rathje (2019).

1. We collected acceleration time series at surface and downhole accelerometers for both records and simulations.

2. For ETF, a 5th-order Butterworth filter with a pass band of 0.5-12 Hz was applied to the demeaned and detrended acceleration records, where signals at frequencies below 0.5 Hz were discarded to minimize the contribution from low-frequency noise interference. I will test whether filtering on synthetics changes the TTF.

3. A 2nd-order polynomial baseline correction was applied to the displacement time histories obtained by integrating the accelerations twice (For synthetics, integration is done once since the output is velocities).   
4. The ETFs and TTFs were obtained as the ratio of the Fourier spectral amplitude between the surface and downhole acceleration time histories for all the events and simulations, respectively.  

5. We further smoothed the TFs using the Konno-Ohmachi smoothing window in the frequency domain (Konno and Ohmachi, 1998).  

Here, we woule like to show results step by step, and investigate whether applying the bandpass filter in step (2) or adding tapering functions to synthetics affects the results, and in consequense explain the sensitivity of the TTFs to source selection. E.g. the TTFs of the same 1D model using various ricker sources:

![](GVDA_ricker_sources.png)

Below is the preparation of calculating TFs. The procesure of computing a TTF is realized by surboutine `comp_TTF`, and also shown step by step, 

```python
# Prepare parameters used later

dt = 0.01  # time step
t_ref = 150  # time series duration
f_lim = 11.2  # upper boundary of frequency
N_half = int(t_ref / dt) // 2  # half length of time steps
ix, iy = 4, 2  # index of the site

model = "TKCH05_lin_3d_B_2v680_bd98_h005l100"
vx = np.fromfile(f"../data/vx_{model}.bin", "f").reshape((-1, 2, 4, 8))  # dt = 0.005s in simulations
vy = np.fromfile(f"../data/vy_{model}.bin", "f").reshape((-1, 2, 4, 8))  # dt = 0.005s

vx = vx[::2, :, iy, ix]  # shape(15000, 2), dt = 0.01s so that consistent with data
vy = vy[::2, :, iy, ix]

t = np.arange(len(vx)) * dt  # time range

f_ref = np.linspace(0, 1.0 / (2.0 * dt), N_half)  # frequency range
sm_ref = calculate_smoothing_matrix(f_ref, normalize=True)  # the smoothing matrix

# If 5s zeros padded in the end
pad_seconds = 5
pad_length = int(pad_seconds / 0.01) // 2
f_pad = np.linspace(0, 1.0 / (2.0 * dt), N_half + pad_length)  # frequency range
sm_pad = calculate_smoothing_matrix(f_pad, normalize=True)  # the smoothing matrix
```

```python
ttf_1d = np.load("../data/ttf_1d_lin.npy")
ttf_1d_med = np.median(ttf_1d, axis=1)
etf = np.load("../data/etf.npy")
print(etf.shape)
etf_up = np.load("../data/etf_up.npy")
etf_low = np.load("../data/etf_low.npy")
etf_med = np.median(etf, axis=1)
```

### Step 1

```python
fig, ax = plt.subplots(2, 1, dpi=600)
fig.tight_layout()
ax[0].plot(t, vx)
ax[1].plot(t, vy)
ax[0].set(xlabel='Time (s)', ylabel='Vx (m/s)')
ax[0].legend(['Surface', 'Downhole'])
ax[1].set(xlabel='Time (s)', ylabel='Vy (m/s)')

print(f"The slope of Vx-surface is: {np.polyfit(t, vx[:, 0], 1)[0]:.4e}")
print(f"The slope of Vx-Downhole is: {np.polyfit(t, vx[:, 1], 1)[0]:.4e}")
print(f"The mean value of Vx-surface in the last 1 second is {np.mean(vx[-100:, 0]):.4e}")
print(f"The mean value of Vx-downholw in the last 1 second is {np.mean(vx[-100:, 1]):.4e}")
pass
```

### Step 2
The figure below shows how the 2nd-order polynomial baseline correction works. 
The displacements are obtained by integrating velocities once, and then trend and mean are removed.

```python
# An example showing how polynomial baseline correction works on x-component
# of surface displacements
dispx = np.cumsum(vx[:, 0], axis=0) * dt
end_dispx = np.mean(dispx[-100:])
print(f"The slope of Disp.X-Surface is: {np.polyfit(t, dispx, 1)[0]:.4e}")
dispx_detrend = polynomial(dispx, order=2, plot=True)
print(f"The mean value of the displacements in the last 1 second was {end_dispx:.4e} initially, "
      f"and {np.mean(dispx_detrend[-100:]):.4e} after baseline correction.")
pass
```

### Step 3
The figure below shows the accelerations after processing, which will be used to compute spectra ratio as the transfer function.

```python
dispx = np.cumsum(vx, axis=0) * dt
for i in range(2):
    dispx[:, i] = polynomial(dispx[:, i], order=2)
accx = np.gradient(np.gradient(dispx, dt, axis=0), dt, axis=0)
dispy = np.cumsum(vy, axis=0) * dt
for i in range(2):
    dispy[:, i] = polynomial(dispy[:, i], order=2)
accy = np.gradient(np.gradient(dispy, dt, axis=0), dt, axis=0)

fig, ax = plt.subplots(2, 1, dpi=600)
ax[0].plot(t, accx, label='accx')
ax[0].set(xlabel='Time (s)', ylabel='Acc. X (m/s/s)')
ax[0].legend(['Surface', 'Downhole'])
ax[1].plot(t, accy, label='accx')
ax[1].set(xlabel='Time (s)', ylabel='Acc. Y (m/s/s)')
ax[1].legend(['Surface', 'Downhole'])

pass
```

### Step 4
The figure below shows the acceleration spectra. No filter is applied.

```python
# This is the process of calculating TTF, without calling the 
fx = np.abs(fft(accx, axis=0) * dt)
print(np.max(fx, axis=0))
fy = np.abs(fft(accy, axis=0) * dt)

fig, ax = plt.subplots(2, 1, dpi=600)
ax[0].loglog(f_ref, fx[1 : N_half + 1, :])
ax[0].set(xlabel='Frequency (Hz)', ylabel='Spec. Acc. X (m/s/s)')
ax[0].legend(['Surface', 'Downhole'])
ax[1].loglog(f_ref, fy[1 : N_half + 1, :])
ax[1].set(xlabel='Frequency (Hz)', ylabel='Spec. Acc. Y (m/s/s)')
ax[1].legend(['Surface', 'Downhole'])
```

### Step 5
We take the ratio between the root mean square of the two horizontal compoennts of the surface and downhole spectra, which is the transfer function. Smoothing is applied in the end.

For every event, the ETF is computed in the same way. The 95% conficence interval of 36 ETFS is plotted in gray.

```python
fa = np.sqrt(fx ** 2 + fy ** 2)  # RMS spectra
N = len(fx)
f_tmp = np.linspace(0, 1 / 2 / dt, N // 2)
tmp = fa[1: N // 2 + 1, 0] / fa[1 : N // 2 + 1, 1]
tmp[np.isnan(tmp)] = 1  # avoid divided-by-zero error
tmp[np.isinf(tmp)] = 1
res = apply_smoothing_matrix(tmp, sm_ref, 1)

fig, ax = plt.subplots(dpi=600)
ax.fill_between(f_ref, etf_low, etf_up, color="gray", alpha=0.3, label="ETF 95% CI")
ax.plot(f_ref, res, label="TTF 3D")
ax.plot(f_ref, ttf_1d_med, label="TTF 1D")
plot_set(ax, 2 * etf_up.max())
```

### Band-pass Filter
The reference TF is calculated without filtering/tapering/baseline correction/detrending velocityes.

If we applied a 0.5 - 12 Hz to the synthetics, the resulting TF shows almost no change.

```python
res_ref = comp_ttf(model, dt, 12, sm_ref)
res_filt = comp_ttf(model, dt, 12, sm_ref, filt=True)
fig, ax = plt.subplots(dpi=600)
ax.fill_between(f_ref, etf_low, etf_up, color="gray", alpha=0.3, label="ETF 95% CI")
ax.plot(f_ref, res_ref, label="TTF 3D ref")
ax.plot(f_ref, res_filt, label="TTF 3D filtered")
plot_set(ax, 2 * etf_up.max())
```

### Taper

Though the synthetics approach zero in the end, tapering might still be helpful. I used a hanning window to taper the last 2 seconds of the syntehtics, then padded 5-second zeros in the end. 

```python
fig, ax = plt.subplots(2, 1, dpi=600)
fig.tight_layout()
vx_taper = taper_vel(vx, dt, 2, 5)  # tapering window = 2s, padding 5s zeros
vy_taper = taper_vel(vy, dt, 2, 5)

t_taper = dt * np.arange(len(vx_taper))
ax[0].plot(t_taper, vx_taper)
ax[0].set(xlabel='Time (s)', ylabel='Vx (m/s)')
ax[0].legend(['Surface', 'Downhole'])
ax[1].plot(t_taper, vy_taper)
ax[1].set(xlabel='Time (s)', ylabel='Vy (m/s)')

print("After adding tapering: \n")
print(f"The slope of Vx-surface is: {np.polyfit(t_taper, vx_taper[:, 0], 1)[0]}")
print(f"The slope of Vx-Downhole is: {np.polyfit(t_taper, vx_taper[:, 1], 1)[0]}")

pass
```

```python
res_taper = comp_ttf(model, dt, 12, sm_pad, taper=2)  # 2s tapering
fig, ax = plt.subplots(dpi=600)
ax.fill_between(f_ref, etf_low, etf_up, color="gray", alpha=0.3, label="ETF 95% CI")
ax.plot(f_ref, res, label="TTF 3D ref")
ax.plot(f_pad, res_taper, label="TTF 3D taper-2s")
plot_set(ax, 2 * etf_up.max())
```

We see that the tapering changes the TTF at frequency above 8.5 hz, but has no effects at lower frequencies.

Also, the length of the tapering window doesn't change the TF in the frequency range we considered. Below is a test where the window length is 10s instead of 2s in the figure above. The difference between two TTF is not visiually discriminable in log scale, at a level of 0.001 - 0.1.

```python
res_taper10 = comp_ttf(model, dt, 12, sm_pad, taper=10)
fig, ax = plt.subplots(dpi=600)
ax.fill_between(f_ref, etf_low, etf_up, color="gray", alpha=0.3, label="ETF 95% CI")
ax.plot(f_ref, res_ref, label="TTF 3D ref")
ax.plot(f_pad, res_taper, label="TTF 3D taper 2s")
ax.plot(f_pad, res_taper10, label="TTF 3D taper 10s")
plot_set(ax, 2 * etf_up.max())
```

### Detrend vel
I used to detrend the velocities for synthetics and data, but it may be not needed in this case.

```python
res_detrend = comp_ttf(model, dt, 12, sm_pad, detrend=True)
fig, ax = plt.subplots(dpi=600)
ax.fill_between(f_ref, etf_low, etf_up, color="gray", alpha=0.3, label="ETF 95% CI")
ax.plot(f_ref, res_ref, label="TTF 3D ref")
ax.plot(f_ref, res_detrend, label="TTF 3D detrend_vel")
plot_set(ax, 2 * etf_up.max())
```

## Summary

It seems the processings including filtering, detrending velocities, baseline correcting displacements are not changing the TTF. The tapering function affects the TF above 8.5 Hz. 

It is not very likely that sensitivity of the TTFs to the source is due to these processing steps.

```python
site_lat, site_lon = 43.1211, 143.6181
N_events = 36  # Only 36 selected

data_dir = "/Users/zhh076/work/preevents/data/TKCH05/"
events = os.listdir(data_dir)
events = sorted([e for e in events if 'txt' not in e])
table = pd.DataFrame(index=np.arange(len(events) // 6), columns=['ID', 'Date', 'Time', 'Mw', 'Latitude', 'Longitude', 'Depth (km)', 'Distance (km)', 'Azimuth'])

def read_TKCH05_event(i, event):
    with open(Path(data_dir, event), 'r') as fid:
        for i in range(9):
            line = fid.readline().split()
            if 'Origin' in line[0]:
                date = line[2]
                time = line[3]
            elif 'Lat' in line[0]:
                lat = float(line[-1])
            elif 'Long' in line[0]:
                lon = float(line[-1])
            elif 'Depth' in line[0]:
                depth = int(line[-1])
            elif 'Mag' in line[0]:
                mag = float(line[-1])

        geod = Geodesic.WGS84.Inverse(lat, lon, site_lat, site_lon)
        azi = int(np.round(geod['azi2'] % 360))  # make sure positive
        dist = geod['s12'] / 1000  # m --> km
    res = [i, date, time, f'{mag:.2f}', f'{lat:.4f}', f'{lon:.4f}', depth, f'{dist:.2f}', azi]
    
    return res


for i in range(len(events) // 6):
    event = events[6 * i]
    table.loc[i] = read_TKCH05_event(i, event)
table = table[(table['Mw'] > '3') & (table['Mw'] < '6.5')]
tmp = table.loc[table['Date'] == '2008/08/29']

table = table.sample(N_events, random_state=128)
table = table.sort_values(by=['Date', 'Time'])
table.iloc[0] = tmp.values
table['Date'] = pd.to_datetime(table['Date']).dt.strftime('%y/%m/%d')
table['ID'] = np.arange(1, N_events + 1)
table = table.rename(columns={"Date": "Date (yy/mm/dd)", "Time": "Time (hh:mm:ss)"})
table.to_csv('/Users/zhh076/work/preevents/results/TKCH05_etf_events.csv', index=False)
```
