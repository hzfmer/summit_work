```python
import subprocess
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import MultipleLocator
from pathlib2 import Path
import csv
import mysql.connector
%matplotlib inline

import seaborn as sns; 
sns.set(context='paper', style='whitegrid', font_scale=1.3, rc={"xtick.bottom" : True, "ytick.left" : True})
sns.set_style({'legend_frameon':False})
```

```python
def distance(lon1, lat1, lon2=-117.932587, lat2=33.918633):
    lat1, lon1, lat2, lon2 = np.radians((lat1, lon1, lat2, lon2))
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    d = 0.5 - np.cos(dlat) / 2 + np.cos(lat1) * np.cos(lat2)  * (1 - np.cos(dlon)) / 2
    return 12742 * np.arcsin(np.sqrt(d))
```

```python
def call_sub(cmd, shell=False, timeout=20):
    p = subprocess.run(cmd, shell=shell, capture_output=True,
                       timeout=timeout, check=True)
#     if p.stderr:
#         print("Error\n", p)
    return p
```

```python
# Start a connection
conn = mysql.connector.connect(
    host="moment.usc.edu",
    user="cybershk_ro",
    password="CyberShake2007",
    database="CyberShake",
    use_unicode=True,
)

# Query site table
cmd = (
    f'select S.CS_Short_Name, S.CS_Site_Name, S.CS_Site_Lat, S.CS_Site_Lon '
    f'from CyberShake_Sites S, CyberShake_Runs R '
    f'where R.Study_ID={STUDY_ID} '
    f'and R.Site_ID=S.CS_Site_ID;'
)
all_sites = pd.read_sql(cmd, conn)
all_sites.columns = ["name", "deccription", "lat", "lon"]

cursor = conn.cursor()
```

```python
def select_by_dist(sites, lon=0, lat=0, site_name="", thresh=10):
    """Select sites within 10 km to (lon, lat)"""
    if site_name:
        site = sites[sites['name'] == site_name]
        lon, lat = site.lon.item(), site.lat.item()
    print(f'Processing {site_name} at lon = {lon:.3f}, lat = {lat:.3f}')
    sites['dist'] = sites.apply(lambda row: distance(row.lon, row.lat, lon, lat), axis=1)
    sites.sort_values('dist', inplace=True)
    return sites[sites['dist'] < thresh]
```

```python
def get_run_id(site_name):
    cmd = (
        f'select R.Run_ID '
        f'from CyberShake_Runs R, CyberShake_Sites S '
        f'where R.Site_ID=S.CS_Site_ID '
        f'and R.Study_ID=7 '
        f'and S.CS_Short_Name={repr(site_name)} '
    )
    run_id = pd.read_sql(cmd, conn)
    return run_id.iloc[0].item()
```

```python
def get_disagg(file_name):
    tmp = pd.read_csv(
        file_name,
        delimiter='\t',
        skiprows=20,
        nrows=1,
    )
    columns = tmp.columns.str.lower().str.replace(' ', '').to_list() + ['total']
    df = pd.read_csv(
        file_name,
        delimiter='\t',
        skiprows=22,
        nrows=110,
        names=columns,
    )
    return df
```

```python
def plot_disagg(df=None, file_name="", title="", ax=None):
    if df is None:
        df = get_disagg(file_name)
    # Specify colors if length known
    colors = mpl.cm.bwr_r(np.linspace(0, 1, len(df.columns) - 3))

    # prepare 3d axes
    if ax is None:
        fig = plt.figure(figsize=(10,6), dpi=300)
        ax = Axes3D(fig)

    N = len(df.columns)
    # thickness of the bars
    dx, dy = 10, .5

    for i, row in df.iterrows():
        # set up positions for the bars 
        if row.iloc[-1] == 0:
            continue
        xpos = [row.iloc[0]] * (N - 3)
        ypos = [row.iloc[1]] * (N - 3)
        dz = row.iloc[2:-1]
        zpos = dz.cumsum().to_list()
        zpos = [0] + zpos[:-1]
        #     # set the ticks in the middle of the bars
        #     ax.set_xticks(xpos + dx/2)
        #     ax.set_yticks(ypos + dy/2)
        # plot 
        ax.bar3d(xpos,ypos,zpos,dx,dy,dz, color=colors)

    # put the column / index labels
    # ax.w_yaxis.set_ticklabels(eg.columns)
    # ax.w_xaxis.set_ticklabels(eg.index)

    # name the axes
    ax.set(
        xlabel = 'Rupture Distance (km)',
        ylabel = 'Magnitude',
        zlabel = '%Contribution',
        xlim = (df['dist'].min(), df['dist'].max()),
        ylim = (df['mag'].min(), df['mag'].max()),
    )
    if title:
        ax.set_title(title)
    ax.xaxis.set_major_locator(MultipleLocator(10.0))
    ax.yaxis.set_major_locator(MultipleLocator(1.0))
    ax.zaxis._axinfo['juggled'] = (1, 2, 0)
    #plt.show()
```

```python
file_name = 'LADT_ERF36_Run4283_DisaggPOE_1.0E-4_SA_0.2sec_RotD50_2019_01_31.txt'
plot_disagg(file_name=file_name, title=f"LADT at 0.2s")
```

```python
def run(lon=-119, lat=34.5, site_name="", period=0.2, thresh=10,
        force_update=0, timeout=20, agg=['min', 'mean', 'max']):
    global all_sites
    sites = select_by_dist(all_sites, lon=lon, lat=lat,
                           site_name=site_name, thresh=thresh)
    
    if site_name:
        lon = np.round(sites[sites['name'] == site_name]['lon'].item(), 2)
        lat = np.round(sites[sites['name'] == site_name]['lat'].item(), 2)
    directory = Path(f'{"loc_" + "_".join([str(lon), str(lat),  str(period)])}')
    print(f'Neaby sites: {sites["name"].to_list()}')
    count = len([directory.glob('*txt')])
    dfs = []
    if not count or force_update:
        for site_name in sites['name']:
            cmd = [
                f'./disagg_plot_wrapper.sh',
                f'--site {site_name}',
                f'--component RotD50',
                f'--period {period}',
                f'--probs 1e-4 --erf-file MeanUCERF.xml --type png,txt',
                f'-o {directory}',
            ]
            try:
                print(call_sub(cmd, timeout=timeout))
            except (subprocess.TimeoutExpired, subprocess.CalledProcessError):
                continue
            print(f'Site {site_name} queried')
    print(directory)
    for f in directory.glob('*txt'):
        file_name = str(f)
        site_name = file_name.split('/')[1].split('_')[0]
        if sites[sites['name']==site_name]['dist'].item() < thresh:
            dfs.append(get_disagg(file_name))
            
    fig = plt.figure(figsize=plt.figaspect(0.33), dpi=300)
    fig.subplots_adjust(wspace=0.5)
    fig.suptitle(f"Disagg {len(dfs)}  sites near ({lon:.2f}, {lat:.2f}) (<{thresh}km) at {period}s")
    ax = fig.add_subplot(1, 3, 1, projection='3d')
    res = pd.concat(dfs).groupby(level=0).min()
    plot_disagg(res, title="Min", ax=ax)
    
    ax = fig.add_subplot(1, 3, 2, projection='3d')
    res = pd.concat(dfs).groupby(level=0).mean()
    plot_disagg(res, title="Mean", ax=ax)
    
    ax = fig.add_subplot(1, 3, 3, projection='3d')
    res = pd.concat(dfs).groupby(level=0).max()
    plot_disagg(res, title="Max", ax=ax)
        
        
    #fig, ax = plt.subplots(1, len(colors), dpi=500, gridspec_kw={'wspace': 0.8})
    x, y = np.linspace(0, 1, 40), np.linspace(0, 1, 40)
    z = np.ones((40, 40))
    for i, color in enumerate(colors):
        ax = fig.add_subplot(position=[0.1 + 0.8 * i/len(colors), -0.06, 0.8/len(colors), 0.06])
        ax.pcolormesh(x, y, z, color=color, shading='auto')
        ax.set_axis_off()
        ax.set_aspect(1/2)
        pos = list(ax.get_position().bounds)
        x_text = pos[0] #+ pos[2] / 3 
        y_text = pos[1] - 1.2
        ax.text(0.5, y_text, dis.columns[i + 2], ha='center', fontsize=12)
    
    return res, len(dfs)

#run(-118, 34)
site_name = "LADT"
dis, nsite = run(site_name="LADT", period=0.2, thresh=10, 
                 force_update=0, timeout=1000)
```

```python
def run_gmt_plot(lon=-119, lat=34.5, site_name="", thresh=10, cmap='inferno'):
    sites = select_by_dist(all_sites, lon=lon, lat=lat,
                           site_name=site_name, thresh=thresh)
    if site_name:
        lon = np.round(sites[sites['name'] == site_name]['lon'].item(), 2)
        lat = np.round(sites[sites['name'] == site_name]['lat'].item(), 2)
    directory = Path(f'{"loc_" + "_".join([str(lon), str(lat),  str(period)])}')
    sites[['lon', 'lat', 'name']].to_csv('sites_location.txt', sep= ' ', header=None, index=None)
    left, right = sites['lon'].min(), sites['lon'].max()
    bot, top = sites['lat'].min(), sites['lat'].max()
    ext = 0.3
    gmt_cmd = (f'range=-R{left-ext}/{right+ext}/{bot-ext}/{top+ext}\n'
               f'model_region=-R{left}/{right}/{bot}/{top}\n'
               f'gmt begin {str(directory)}/sites_location png\n'
               f'gmt set MAP_FRAME_TYPE plain\n'
               f'gmt set FORMAT_FLOAT_OUT %.12lg\n'
               f'gmt basemap -JM10c $range -Baf -BSWen\n'
               f'gmt grdimage -C @earth_relief_15s -I+d\n'
               #f'gmt blockmean results/{fname}.txt $model_region -I128e > results/temp.txt\n'
               #f'gmt surface results/temp.txt $model_region -Ginterp.grd -C2 -I{radius}k/{radius}k -T0.25\n'
               #f'gmt grdsample interp.grd $model_region -Ginterp.grd -I128e/128e\n'
               #f'gmt grdclip interp.grd -Gclip.grd -Sa{vmax - 0.001 * abs(vmax)}/{vmax - 0.001 * abs(vmax)} -Sb{vmin + 0.001 * abs(vmin)}/{vmin + 0.001 * abs(vmin)}\n'
               #f'gmt makecpt -C{cmap} {"-Ic" if cmap_r else ""} -T{vmin}/{vmax}/{(vmax-vmin)/100}\n'
               #f'gmt grdimage $model_range clip.grd -t40\n'
               f'gmt colorbar -DjMR+o-1c/0+m -I0.3 -G0/NaN -Bx+l"Elevation (m)"\n'
               f'gmt plot sites_location.txt -St0.4c -Gcyan -W0.2p,black\n'
               f'echo {lon} {lat} | gmt plot -Sa0.6c -Gred\n'
               f'gmt coast -JM10c -Swhite\n'
               f'# gmt plot borders.txt -W3p,lightblue -L\n'
               f'# gmt plot -W2p,black -L << EOF\n'
               f'# {left} {bot}\n'
               f'# {right} {bot}\n'
               f'# {right} {top}\n'
               f'# {left} {top}\n'
               f'# EOF\n'
               f'gmt end')
    with open('plot_gof.gmt', 'w') as fid_gmt:
        fid_gmt.write(gmt_cmd)
    cmd = ['bash', 'plot_gof.gmt']
    call_sub(cmd, shell=False)
    
run_gmt_plot(site_name='LADT')
```
